```python
evalue_tresh = 0.04
results_blast= open('prot_blast_swiss.xml')
blast_records = NCBIXML.read(results_blast)
for alignment in blast_records.alignments:
    print(alignment.hit_def)
    for hsp in alignment.hsps:
        if hsp.expect <= evalue_tresh:
            print("        ***ALINHAMENTO***")
            print(f"Sequence: ", alignment.title)
            print("Query " + hsp.query[0:100] + "...")
            print("Match " + hsp.match[0:100] + "...")
            print("Sbjct " + hsp.sbjct[0:100] + "...")
            print()
```
